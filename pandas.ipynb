```python
country_use_python= country_grp['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
country_use_python
```

```python
python_df= pd.concat([country_respondents,country_use_python],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'country':'NumRespondents','LanguageWorkedWith':'NumKnowsPython'},inplace=True)
python_df
```

```python
python_df['PctKnowsPython']=(python_df['NumKnowsPython']/python_df['Country'])*100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython',ascending=False,inplace=True)
python_df
```
